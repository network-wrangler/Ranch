```python
# join conflation database with input database

conflation_df = pd.merge(
    conflation_df.rename(columns = {'pp_unique_id' : 'unique_id'}),
    county_df[['unique_id', 'BASE_LANES']],
    how = 'left',
    on = 'unique_id'
)

conflation_df = conflation_df[conflation_df['BASE_LANES'] > 0]
```

```python
conflation_df.info()
```

```python
# drop duplciates in the conflation database

conflation_df.drop_duplicates(inplace = True)

conflation_df.info()
```

```python
# in conflation df, aggregate based on shstReferenceId, get all number of lanes for each shstReferenceId

lanes_conflation_df = conflation_df.groupby(
    ['shstReferenceId']
)['BASE_LANES'].apply(list).to_frame().reset_index()
```

```python
lanes_conflation_df['base_lanes_min'] = lanes_conflation_df['BASE_LANES'].apply(lambda x: min(set(x)))
lanes_conflation_df['base_lanes_max'] = lanes_conflation_df['BASE_LANES'].apply(lambda x: max(set(x)))
```

```python
lanes_conflation_df[lanes_conflation_df.base_lanes_min != lanes_conflation_df.base_lanes_max]
```

```python
lanes_conflation_df = pd.merge(
    roadway_network.links_df[['shstReferenceId']],
    lanes_conflation_df,
    how = 'left',
    on = 'shstReferenceId'
)
```

```python
lanes_conflation_df[(lanes_conflation_df.base_lanes_min != lanes_conflation_df.base_lanes_max) &
                    (lanes_conflation_df.BASE_LANES.notnull())
                    ]
```

```python
lanes_conflation_df[lanes_conflation_df.shstReferenceId == '5c6836b993450e870c8beaad7e32dfc5']
```

```python
lanes_conflation_df.to_csv(os.path.join(output_dir, 'sjmodel_legacy_lanes.csv'), index = False)
```
